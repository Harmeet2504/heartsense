```python
#Import dependencies
import os
import pandas as pd
import json
import numpy as np
import requests
```

```python
#Loading csv
cleveland_df= pd.read_csv('Cleveland_data.csv')
cleveland_df.head()
```

```python
#Check data type of each column
cleveland_df.dtypes
```

```python
#Age: displays the age of the individual.
# Sex: displays the gender of the individual using the following format :
# 1 = male
# 0 = female
# Chest-pain type: displays the type of chest-pain experienced by the individual using the following format :
# 1 = typical angina
# 2 = atypical angina
# 3 = non — anginal pain
# 4 = asymptotic
# Resting Blood Pressure: displays the resting blood pressure value of an individual in mmHg (unit)
# Serum Cholestrol: displays the serum cholesterol in mg/dl (unit)
# Fasting Blood Sugar: compares the fasting blood sugar value of an individual with 120mg/dl.
# If fasting blood sugar > 120mg/dl then : 1 (true)
# else : 0 (false)
# Resting ECG : displays resting electrocardiographic results
# 0 = normal
# 1 = having ST-T wave abnormality
# 2 = left ventricular hyperthrophy
# Max heart rate achieved : displays the max heart rate achieved by an individual.
# Exercise induced angina :
# 1 = yes
# 0 = no
# ST depression induced by exercise relative to rest: displays the value which is an integer or float.
# Peak exercise ST segment :
# 1 = upsloping
# 2 = flat
# 3 = downsloping
# ca: Number of major vessels (0–3) colored by flourosopy : displays the value as integer or float.
# Thal : displays the thalassemia :
# 3 = normal
# 6 = fixed defect
# 7 = reversible defect
# Diagnosis of heart disease : Displays whether the individual is suffering from heart disease or not :
# 0 = absence
# 1, 2, 3, 4 = present.
```

```python
#Rename values of "sex" column
cleveland_df.loc[(cleveland_df.sex == '1'),'sex']='Male'
cleveland_df.loc[(cleveland_df.sex == 0),'sex']='Female'
cleveland_df
```

```python
#Rename values of "cp" column
cleveland_df.loc[(cleveland_df.cp == 1),'cp']='Typical angina'
cleveland_df.loc[(cleveland_df.cp == 2),'cp']='Atypical angina'
cleveland_df.loc[(cleveland_df.cp == 3),'cp']='Non — anginal pain'
cleveland_df.loc[(cleveland_df.cp == 4),'cp']='Asymptotic'
cleveland_df
```

```python
#Rename values of "fbs" column
cleveland_df.loc[(cleveland_df.fbs == 1),'fbs']='> 120mg/dl'
cleveland_df.loc[(cleveland_df.fbs == 0),'fbs']='< 120mg/dl'
cleveland_df
```

```python
#Rename values of "restecg" column
cleveland_df.loc[(cleveland_df.restecg == 0),'restecg']='Normal'
cleveland_df.loc[(cleveland_df.restecg == 1),'restecg']='ST-T wave abnormality'
cleveland_df.loc[(cleveland_df.restecg == 2),'restecg']='Left ventricular hyperthrophy'
cleveland_df
```

```python
#Rename values of "exang" column
cleveland_df.loc[(cleveland_df.exang== 1),'exang']='Yes'
cleveland_df.loc[(cleveland_df.exang == 0),'exang']='No'
cleveland_df
```

```python
#Rename values of "slope" column
cleveland_df.loc[(cleveland_df.slope == 1),'slope']='Upsloping'
cleveland_df.loc[(cleveland_df.slope == 2),'slope']='Flat'
cleveland_df.loc[(cleveland_df.slope == 3),'slope']='Downsloping'
cleveland_df
```

```python
#Rename values of "thal" column
cleveland_df.loc[(cleveland_df.thal == 1),'thal']='Normal'
cleveland_df.loc[(cleveland_df.thal == 2),'thal']='Fixed Defect'
cleveland_df.loc[(cleveland_df.thal == 3),'thal']='Reversible Defect'
cleveland_df
```

```python
cleveland_df.to_csv('cleveland_dataset.csv', index=False)
```
